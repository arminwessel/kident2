```python
import pandas as pd
import numpy as np
import datetime
```

```python
trajfile = 'traj.csv'
df = pd.read_csv(trajfile)
df
```

```python
def get_random_rows(num, A):
    return A[np.random.choice(A.shape[0], num, replace=False)]
```

```python
random_poses_invkin = '5000_random_poses_invkin.csv'
random_poses_invkin = pd.read_csv(random_poses_invkin)
qs = [f'q{i+1}' for i in range(7)]
hundred_random_poses = random_poses_invkin[qs].to_numpy()
hundred_random_poses = get_random_rows(100, hundred_random_poses)
hundred_random_poses = hundred_random_poses.T
np.shape(hundred_random_poses)
```

```python
current_datetime = datetime.datetime.now()
timestamp = current_datetime.timestamp()
timestamp = datetime.datetime.fromtimestamp(timestamp).strftime("%Y-%m-%d-%H-%M-%S")

df_export = pd.DataFrame(hundred_random_poses)
df_export.to_csv('random_traj_'+timestamp+'.csv')
```
